```python
import pandas as pd
import zipfile
import kaggle
import csv
```

```python
financial_data = 'financial-statements-of-major-companies2009-2023.zip'
with zipfile.ZipFile(financial_data, 'r') as file:
    file.extractall()

f = open('Financial Statements.csv', 'r')
reader = csv.reader(f)
data_info = next(reader)
```

```python
#sorting out microsoft and google finance data.
with open('google_finance_data.csv', 'w', newline ='') as g_file:
    with open('microsoft_finance_data.csv', 'w', newline ='') as m_file:
        writer1 = csv.writer(g_file)
        writer2 = csv.writer(m_file)
        writer1.writerow(data_info)
        writer2.writerow(data_info)
        for row in reader:
            if row[1] == 'GOOG':
                writer1.writerow(row)
            elif row[1] == 'MSFT':
                writer2.writerow(row)
```

```python
google_data = pd.read_csv("google_finance_data.csv")
microsoft_data = pd.read_csv("microsoft_finance_data.csv")
all_data = pd.read_csv("Financial Statements.csv")
```

```python
#replacing spaces in column names w '_' and making all lowercase
google_data.columns = [g.lower().replace(' ', '_').replace('(', '_').replace(')', '_') for g in google_data.columns]
microsoft_data.columns = [m.lower().replace(' ', '_').replace('(', '_').replace(')', '_')  for m in microsoft_data.columns]
all_data.columns = [a.lower().replace(' ', '_').replace('(', '_').replace(')', '_')  for a in all_data.columns]

#viewing new df
all_data
```

```python
new_cols_dict = {
'year' : 'year',
 'company_' : 'company',
 'category' : 'category',
 'market_cap_in_b_usd_' : 'market_capital_in_billions_usd',
 'revenue': 'revenue_in_millions_usd',
 'gross_profit' : 'gross_profit_in_millions_usd',
 'net_income' : 'net_income_in_millions_usd',
 'earning_per_share' : 'earning_per_share_in_usd' ,
 'ebitda' : 'ebitda_in_millions_usd',
 'share_holder_equity' :'share_holder_equity_in_millions_usd',
 'cash_flow_from_operating' : 'cash_flow_from_operating_in_millions_usd',
 'cash_flow_from_investing' : 'cash_flow_from_investing_in_millions_usd',
 'cash_flow_from_financial_activities' : 'cash_flow_from_financial_activities_in_millions_usd',
 'current_ratio' : 'current_ratio',
 'debt/equity_ratio' : 'debt/equity_ratio',
 'roe' : 'return_on_equity_percentage',
 'roa' : 'return_on_asset_percentage',
 'roi' : 'return_on_investment_percentage',
 'net_profit_margin' : 'net_profit_margin_percentage',
 'free_cash_flow_per_share' : 'free_cash_flow_per_share_in_usd',
 'return_on_tangible_equity' : 'return_on_tangible_equity_percentage',
 'number_of_employees' : 'number_of_employees' ,
 'inflation_rate_in_us_' : 'inflation_rate_in_us'
}

all_data.rename(new_cols_dict, axis=1, inplace=True)
microsoft_data.rename(new_cols_dict, axis=1, inplace=True)
google_data.rename(new_cols_dict, axis=1, inplace=True)
```

```python
data_frames = [all_data, google_data, microsoft_data]

# List of columns to transform to percentages
columns_to_transform = [
    'return_on_equity_percentage',
    'return_on_asset_percentage',
    'return_on_investment_percentage',
    'net_profit_margin_percentage',
    'return_on_tangible_equity_percentage',
]

# Loop through each data frame and apply the transformation
for data_frame in data_frames:
    for column in columns_to_transform:
        if column in data_frame.columns and data_frame[column].dtype == 'float64':
            data_frame[column] /= 100

#check
microsoft_data
```

```python
company_dict = {
    'AAPL' : 'Apple',
    'GOOG' : 'Google',
    'MSFT' : 'Mircosoft',
    'NVDA' : 'Nvidia',
    'PYPL' : 'PayPal',
    'AIG' : 'American International Group',
    'PCG' : 'Pacific Gas and Electric',
    'SHLDQ' : 'Sears Holdings Corp',
    'MCD' : 'McDonalds',
    'BCS' : 'Barclays',
    'INTC' : 'Intel',
    'AMZN' : 'Amazon'
}
category_dict = {
    'BANK' : 'Bank',
    'FOOD' : 'Food',
    'ELEC' : 'Electric',
    'LOGI' : 'Logi'
}

all_data['company'] = all_data['company'].map(lambda x: company_dict.get(x, x))
all_data['category'] = all_data['category'].map(lambda x: category_dict.get(x, x))
microsoft_data['company'].replace(company_dict, inplace=True)
google_data['company'].replace(company_dict, inplace=True)

#check
all_data.head
```
